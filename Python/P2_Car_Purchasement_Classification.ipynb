 #### Feature Engineering

```python
data = pd.get_dummies(data=data, drop_first=False)
data.head()
```

```python
data = data.drop('Gender_Male', axis='columns')
```

```python
X = data[['AnnualSalary','Age', 'Gender_Female']].copy()
```

```python
y = data['Purchased'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=69)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def model_performance(predictions):
  print('Accuracy score in model is {}'.format(accuracy_score(y_test, predictions)))
```

#### Machine Learning

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train,y_train.values.ravel())
log_predictions = log_model.predict(scaled_X_test)
model_performance(log_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
test_errors = []
for k in range(1,30):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(scaled_X_train, y_train.values.ravel()) 
  knn_pred = knn_model.predict(scaled_X_test)
  test_error_rate = 1 - accuracy_score(y_test, knn_pred)
  test_errors.append(test_error_rate)
```

```python
plt.plot(range(1,30), test_errors)
plt.ylabel('Error Rate')
plt.xlabel('K Neighbors')
plt.show()
```

```python
knn_elbow_method = KNeighborsClassifier(n_neighbors=12)
knn_elbow_method.fit(scaled_X_train, y_train.values.ravel())
knn_pred = knn_model.predict(scaled_X_test)
model_performance(knn_pred)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
knn = KNeighborsClassifier()
k_values = list(range(1,30))
pipe = Pipeline(steps = [('knn', knn)])
param_grid = {'knn__n_neighbors': k_values}
cv_classifier = GridSearchCV(pipe, param_grid, cv=5, scoring= 'accuracy')
```

```python
cv_classifier.fit(scaled_X_train, y_train.values.ravel())
print(cv_classifier.best_params_)
```

```python
pred_grid_knn = cv_classifier.predict(scaled_X_test)
model_performance(pred_grid_knn)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {'C':[.01,.1,.5,1],
                  'kernel':['linear','rbf','poly']
                  }
grid_svr = GridSearchCV(svm, param_grid_svr)
```

```python
grid_svr.fit(scaled_X_train, y_train.values.ravel())
pred_svr = grid_svr.predict(scaled_X_test)
model_performance(pred_svr)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(scaled_X_train, y_train.values.ravel())
```

```python
tree_pred = tree_model.predict(scaled_X_test)
model_performance(tree_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr_model = RandomForestClassifier()
n_estimators = [32,64,128,256]
max_features = [2,3,4]
bootstrap= [True, False]
oob_score= [True, False]
param_grid_rfr = {'n_estimators':n_estimators,
                  'max_features':max_features,
                  'bootstrap':bootstrap,
                  'oob_score':oob_score
                  }
grid_rfr = GridSearchCV(rfr_model,param_grid_rfr)
grid_rfr.fit(scaled_X_train, y_train.values.ravel())
```

```python
print(grid_rfr.best_params_)
```

```python
rfc = RandomForestClassifier(max_features=3, 
                             n_estimators=64,
                             oob_score=True,
                             bootstrap=True)
```

```python
rfc.fit(scaled_X_train, y_train.values.ravel())
pred_rfc = rfc.predict(scaled_X_test)
model_performance(pred_rfc)
```
